```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans

from google.colab import files
uploaded = files.upload()
import io
data = pd.read_csv(io.BytesIO(uploaded['ClientesKMeans.csv']))
data

#ploteo de datos
plt.scatter (data['Satisfacci√≥n'], data['Lealtad'])
plt.xlabel('Satisfacci√≥n')
plt.ylabel('Lealtad')
plt.show()

#seleccion de caracteristicas
x = data.copy()
kmeans = KMeans(2)
kmeans.fit(x)

#Resultados y su ploto
g = x.copy()
g['p'] = kmeans.fit_predict(x)
print(g)
plt.scatter(g['Satisfacci√≥n'],g['Lealtad'],c=g['p'],cmap='rainbow')
plt.xlabel('Satisfacci√≥n')
plt.ylabel('Lealtad')
plt.show()

#Normalizacion
from sklearn import preprocessing
xNorm = preprocessing.scale(x)
print(xNorm)

#M√©todo del codo
codo = []
for i in range(1,30):
  kmeans = KMeans(i)
  kmeans.fit(xNorm)
  codo.append(kmeans.inertia_)
print(codo)
num=range(1,30)
plt.scatter(codo,num)
plt.xlabel('Error Cuadratico')
plt.ylabel('Numero')
plt.show()

#seleccion de caracteristicas
x = data.copy()
kmeans = KMeans(4)
kmeans.fit(x)

#Resultados y su ploto
g = x.copy()
g['p'] = kmeans.fit_predict(x)
print(g)
plt.scatter(g['Satisfacci√≥n'],g['Lealtad'],c=g['p'],cmap='rainbow')
plt.xlabel('Satisfacci√≥n')
plt.ylabel('Lealtad')
plt.show()
```
